#### Провести интерполяционный полином $P_n(x)$ через точки $x_k = 1 + \frac{k}{n}$, $y_k = \ln{x_k}$ при $k = 0, ...,n$, $n = 4,...,15$

#### Нарисовать графики $P_n(x) - \ln{x}$

```python
import matplotlib.pyplot as plt
import math
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

%matplotlib inline
```

 $f(x) = \sum_{i=0}^{n}{a_in_i(x)}$, где  $n_i(x) = \prod_{j=0}^{i-1}(x-x_j)$

$ f[x_k, x_{k-1}, \dots, x_{1}, x_0] = \frac{f[x_k, x_{k-1}, \dots, x_{2}, x_2] - f[x_{k-1}, x_{k-2}, \dots, x_{1}, x_0]}{x_k-x_0}$

```python
def get_x(n):
    x_list = []
    for k in range(n + 1):
        x_list.append(1 + (k/n))
    return x_list

def get_y(x_list):
    y_list = []
    for x in x_list:
        y_list.append(np.log(x)) 
    return y_list
```

```python
def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
            
    return coef

def newton_poly(coef, x_data, x):
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n - k] + (x - x_data[n - k]) * p
    return p
```

```python
def calc(x_new, y_new):
    res = []
    for i in range(len(x_new)):
        res.append(abs(y_new[i] - np.log(x_new[i])))
    return res
```

```python
def result(n):
    x = get_x(n)
    y = get_y(x)
    a_s = divided_diff(x, y)[0, :]
    x_new = np.arange(1, 2, 0.003)
    y_new = newton_poly(a_s, x, x_new)

    fig = go.Figure()
    return [x_new, calc(x_new, y_new), max(calc(x_new, y_new)), n]
```

```python
fig = go.Figure()
for i in range(4,25):
    fig.add_trace(go.Scatter(x=result(i)[0], y=result(i)[1],name =f'Newton’s Polynomial Interpolation {i}'))
fig.show()
```
